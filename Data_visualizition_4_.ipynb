Tanh Curve

```python
#tanh curve
import numpy as np
import matplotlib.pyplot as plt
ini=np.linspace(-np.pi, np.pi, 12)
out=np.tanh(ini)
print('in array',ini)
print('out array',out)

plt.plot(ini, out, color='red', marker='o')
plt.title('numpy.tanh()')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

sigmoid curve

```python
#sigmoid curve
#import matplotlib,numpy and math
import matplotlib.pyplot as plt
import numpy as np
import math

x=np.linspace(-100,100,200)
z=1/(1+np.exp(-x))

plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('sigmoid(x)')

plt.show()
```
